We are given `two json` and `one text` file as a raw data.
There are two ways to merge data:
1. Convert all files to csv and then merge
2. merge all files then convert it to csv.

```python
# importing JSON module
import json
```

```python
# open BEHAVIOUR.JSON file (file should be in same directory as file)

with open("behaviour.json",'r') as f1:
   data_1 = json.load(f1)

data_1
```

After seeing data in `behaviour.json` file we realized that the data is in format of ***DICTIONARY***, so first we have to convert it into a ***LIST*** for further computations

```python
# Initailize an empty list to store data of behavior.json
rows = []

# Iterate over each dictionary in the list
for items in data_1:
# Extract the Id and values from the dictionary
    id_ = list(items.keys())[0]
    values = items[id_]

# Create a dictionary with Id as a key and value
row = {'Id':id_}
row.update(values)
rows.append(row)

# Creating DataFrame
df1 = pd.DataFrame(rows)
```

```python
# Displaying DataFrame
print(df1.head())
```

```python
# Removing `ID_` prefix from `Id` column in behaviour.json
df1["Id"] = df1['Id'].str.replace('ID_','')
print(df1.head())
```
